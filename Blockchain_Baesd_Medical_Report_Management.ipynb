```python
# ===== STEP 4: BLOCKCHAIN LINKING LOGIC =====

# Add a column to store the previous record's hash (only run once)
try:
    cursor.execute("ALTER TABLE medical_records ADD COLUMN Previous_Hash TEXT")
    conn.commit()
    print("Previous_Hash column added.")
except sqlite3.OperationalError:
    print("Previous_Hash column already exists.")

# Fetch all records in order of Patient_ID
cursor.execute("SELECT Patient_ID, Record_Hash FROM medical_records ORDER BY Patient_ID ASC")
all_records = cursor.fetchall()

previous_hash = "0" * 64  # Genesis block — first record links to this placeholder

for patient_id, record_hash in all_records:
    # Update this record's Previous_Hash to link to the one before it
    cursor.execute(
        "UPDATE medical_records SET Previous_Hash = ? WHERE Patient_ID = ?",
        (previous_hash, patient_id)
    )
    # This record's hash becomes the "previous hash" for the next one
    previous_hash = record_hash

conn.commit()
print("✅ Blockchain linking complete — every record now references the previous record's hash.")

# Verify by viewing first 5 linked records
import pandas as pd
df_check = pd.read_sql_query(
    "SELECT Patient_ID, Record_Hash, Previous_Hash FROM medical_records LIMIT 5", conn
)
df_check
```

```python
# ===== STEP 5: TAMPER DETECTION FUNCTION =====

def verify_chain():
    cursor.execute("""
        SELECT Patient_ID, Patient_Name, Age, Gender, Disease, Doctor_Name,
               Hospital_Name, Visit_Date, Medicine, Timestamp,
               Record_Hash, Previous_Hash
        FROM medical_records ORDER BY Patient_ID ASC
    """)
    all_rows = cursor.fetchall()

    expected_previous_hash = "0" * 64
    tampered_records = []

    for row in all_rows:
        (patient_id, name, age, gender, disease, doctor, hospital,
         visit_date, medicine, timestamp, stored_hash, stored_previous_hash) = row

        # Recompute hash from current data (same formula used originally)
        raw_string = f"{name}{age}{gender}{disease}{doctor}{hospital}{visit_date}{medicine}{timestamp}"
        recalculated_hash = hashlib.sha256(raw_string.encode()).hexdigest()

        # Check 1: Does stored hash match recalculated hash? (data tampering check)
        if recalculated_hash != stored_hash:
            tampered_records.append((patient_id, "Data was modified"))

        # Check 2: Does this record's Previous_Hash match the expected chain?
        if stored_previous_hash != expected_previous_hash:
            tampered_records.append((patient_id, "Chain link broken"))

        expected_previous_hash = stored_hash

    if not tampered_records:
        print("✅ Blockchain verified — no tampering detected. All", len(all_rows), "records are intact.")
    else:
        print("⚠️ Tampering detected in", len(tampered_records), "record(s):")
        for pid, reason in tampered_records:
            print(f"   - Patient_ID {pid}: {reason}")

    return tampered_records

# Run verification
verify_chain()
```

```python
# ===== STEP 6: AUTHORIZED UPDATE FUNCTION (Proper way to update records) =====

def authorized_update(patient_id, field_name, new_value):
    """
    Properly updates a record and recalculates its hash,
    so it passes blockchain verification (unlike raw tampering).
    """
    # Step 1: Update the specified field
    cursor.execute(f"""
        UPDATE medical_records
        SET {field_name} = ?, Status = 'Updated', Timestamp = ?
        WHERE Patient_ID = ?
    """, (new_value, datetime.now().strftime('%Y-%m-%d %H:%M:%S'), patient_id))
    conn.commit()

    # Step 2: Fetch the updated record's fields (needed to recompute hash)
    cursor.execute("""
        SELECT Patient_Name, Age, Gender, Disease, Doctor_Name,
               Hospital_Name, Visit_Date, Medicine, Timestamp
        FROM medical_records WHERE Patient_ID = ?
    """, (patient_id,))
    name, age, gender, disease, doctor, hospital, visit_date, medicine, timestamp = cursor.fetchone()

    # Step 3: Recalculate the hash with new data
    raw_string = f"{name}{age}{gender}{disease}{doctor}{hospital}{visit_date}{medicine}{timestamp}"
    new_hash = hashlib.sha256(raw_string.encode()).hexdigest()

    # Step 4: Save the new hash
    cursor.execute("UPDATE medical_records SET Record_Hash = ? WHERE Patient_ID = ?",
                   (new_hash, patient_id))
    conn.commit()

    print(f"✅ Patient_ID {patient_id} updated properly. New hash generated.")

# Use it to fix Patient_ID 5 (properly this time)
authorized_update(5, "Disease", "Diabetes")

# Verify again
verify_chain()
```

```python
# ===== STEP 7: PROPER CORRECTION METHOD — ADD AS NEW RECORD =====

def add_correction_record(original_patient_id, field_name, new_value):
    """
    Instead of editing an old record, this creates a NEW record
    that references the original one — just like real blockchain systems.
    """
    # Fetch the original record
    cursor.execute("SELECT * FROM medical_records WHERE Patient_ID = ?", (original_patient_id,))
    columns = [desc[0] for desc in cursor.description]
    original = dict(zip(columns, cursor.fetchone()))

    # Update just the corrected field
    original[field_name] = new_value
    original['Status'] = 'Updated'
    original['Timestamp'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    # Get the current last hash in the chain (to link properly)
    cursor.execute("SELECT Record_Hash FROM medical_records ORDER BY Patient_ID DESC LIMIT 1")
    last_hash = cursor.fetchone()[0]

    # Recalculate hash for this new corrected record
    raw_string = f"{original['Patient_Name']}{original['Age']}{original['Gender']}{original['Disease']}{original['Doctor_Name']}{original['Hospital_Name']}{original['Visit_Date']}{original['Medicine']}{original['Timestamp']}"
    new_hash = hashlib.sha256(raw_string.encode()).hexdigest()

    # Insert as a brand-new record (chain grows, nothing is overwritten)
    cursor.execute('''
        INSERT INTO medical_records (
            Patient_Name, Age, Gender, Blood_Group, Disease, Doctor_Name,
            Hospital_Name, Visit_Date, Medicine, Test_Result, Record_Hash,
            Block_ID, Timestamp, Status, Previous_Hash
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (
        original['Patient_Name'], original['Age'], original['Gender'], original['Blood_Group'],
        original['Disease'], original['Doctor_Name'], original['Hospital_Name'],
        original['Visit_Date'], original['Medicine'], original['Test_Result'],
        new_hash, original['Block_ID'] + 1, original['Timestamp'], original['Status'], last_hash
    ))
    conn.commit()
    print(f"✅ Correction added as a NEW record, linked to original Patient_ID {original_patient_id}.")

# Example: correct Patient_ID 6's disease properly
add_correction_record(6, "Disease", "Hypertension")

# Verify — old records may still show old issues, but the chain grows correctly from here
verify_chain()
```

```python
# ===== STEP 8: SEARCH / LOOKUP FUNCTION =====

def search_patient(patient_id=None, name=None):
    """
    Search medical records by Patient_ID or Patient_Name.
    """
    if patient_id:
        query = "SELECT * FROM medical_records WHERE Patient_ID = ?"
        params = (patient_id,)
    elif name:
        query = "SELECT * FROM medical_records WHERE Patient_Name LIKE ?"
        params = (f"%{name}%",)
    else:
        print("Please provide a Patient_ID or Name to search.")
        return

    df_result = pd.read_sql_query(query, conn, params=params)

    if df_result.empty:
        print("No matching records found.")
    else:
        print(f"Found {len(df_result)} record(s):")
    return df_result

# Example searches
search_patient(patient_id=6)
```

```python
# ===== STEP 9: MENU-DRIVEN USER INTERFACE =====

def add_new_record():
    print("\n--- Add New Patient Record ---")
    name = input("Patient Name: ")
    age = int(input("Age: "))
    gender = input("Gender (Male/Female): ")
    blood_group = input("Blood Group: ")
    disease = input("Disease: ")
    doctor = input("Doctor Name: ")
    hospital = input("Hospital Name: ")
    visit_date = datetime.now().strftime('%Y-%m-%d')
    medicine = input("Medicine: ")
    test_result = input("Test Result: ")
    timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    status = "Active"

    # Get last hash in chain
    cursor.execute("SELECT Record_Hash FROM medical_records ORDER BY Patient_ID DESC LIMIT 1")
    result = cursor.fetchone()
    last_hash = result[0] if result else "0"*64

    cursor.execute("SELECT MAX(Block_ID) FROM medical_records")
    last_block = cursor.fetchone()[0] or 0

    raw_string = f"{name}{age}{gender}{disease}{doctor}{hospital}{visit_date}{medicine}{timestamp}"
    record_hash = hashlib.sha256(raw_string.encode()).hexdigest()

    cursor.execute('''
        INSERT INTO medical_records (
            Patient_Name, Age, Gender, Blood_Group, Disease, Doctor_Name,
            Hospital_Name, Visit_Date, Medicine, Test_Result, Record_Hash,
            Block_ID, Timestamp, Status, Previous_Hash
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (name, age, gender, blood_group, disease, doctor, hospital,
          visit_date, medicine, test_result, record_hash,
          last_block + 1, timestamp, status, last_hash))
    conn.commit()
    print(f"✅ New record added for {name} and linked to blockchain.")


def view_last_10():
    df = pd.read_sql_query(
        "SELECT Patient_ID, Patient_Name, Age, Disease, Doctor_Name, Status FROM medical_records ORDER BY Patient_ID DESC LIMIT 10",
        conn
    )
    print("\n--- Last 10 Records ---")
    print(df.to_string(index=False))


def search_menu():
    print("\n--- Search Patient ---")
    choice = input("Search by (1) Patient_ID or (2) Name? Enter 1 or 2: ")
    if choice == "1":
        pid = int(input("Enter Patient_ID: "))
        result = search_patient(patient_id=pid)
    else:
        name = input("Enter Name (or part of it): ")
        result = search_patient(name=name)
    if result is not None and not result.empty:
        print(result.to_string(index=False))


def verify_menu():
    print("\n--- Verifying Blockchain ---")
    verify_chain()


def correction_menu():
    print("\n--- Add Correction Record ---")
    pid = int(input("Enter Patient_ID to correct: "))
    field = input("Which field to correct? (e.g. Disease, Medicine, Doctor_Name): ")
    new_value = input(f"New value for {field}: ")
    add_correction_record(pid, field, new_value)


def main_menu():
    while True:
        print("\n" + "="*50)
        print(" BLOCKCHAIN BASED MEDICAL RECORD MANAGEMENT SYSTEM")
        print("="*50)
        print("1. Add New Patient Record")
        print("2. Search Patient Record")
        print("3. View Last 10 Records")
        print("4. Verify Blockchain Integrity")
        print("5. Add Correction to a Record")
        print("6. Exit")
        choice = input("Enter your choice (1-6): ")

        if choice == "1":
            add_new_record()
        elif choice == "2":
            search_menu()
        elif choice == "3":
            view_last_10()
        elif choice == "4":
            verify_menu()
        elif choice == "5":
            correction_menu()
        elif choice == "6":
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1-6.")

# Run the menu
main_menu()
```
