```python
import pandas as pd
import numpy as np
import difflib 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
#loading the data from the csv file to a pandas data frame
df=pd.read_csv(r"C:\Users\acer\Downloads\movie dataset\tmdb_5000_movies.csv")
df
```

```python
#printing frst 5 rows of the df
df.head()
```

```python
#num of rows and columns in df
df.shape
```

```python
#selected features
selected_features=['genres','keywords','tagline',]
print(selected_features)
```

```python
df.info()
```

```python
#combining all the file selected feature
combined_features=df['genres']+' '+df['keywords']
```

```python
print(combined_features)
```

```python
#converting the text to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#cosine similarity,getting the similarity scores using cosine similarity
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity.shape)
```

```python
#getting the movie name by user
movie_name=input("Enter your favourite movie name:")
```

```python
#create a list with all the movie names given in the dataset
list_of_all_titles=df['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match 
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#find the index of the movie
index_of_the_movie=df[df.title==close_match].index[0]
print(index_of_the_movie)
```

```python
#getting a list of similar movies
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on their similarity score
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#print the similar movies based on the index
print("Movies suggested for you:\n ")
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if(i<21):
        print(i,'.',title_from_index)
        i+=1
        
```

```python
#MOVIES RECOMMENDATION SYSTEM

movie_name=input("Enter your favourite movie name:")

list_of_all_titles=df['title'].tolist()

find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)

close_match=find_close_match[0]

index_of_the_movie=df[df.title==close_match].index[0]

similarity_score=list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)

print("Movies suggested for you:\n ")
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if(i<21):
        print(i,'.',title_from_index)
        i+=1
```
